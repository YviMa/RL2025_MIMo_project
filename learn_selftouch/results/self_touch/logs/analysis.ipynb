```python
import pickle
import matplotlib.pyplot as plt

with open('training.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
left_ids = set()
right_ids = set()
left_cumulative = []
right_cumulative = []

for episode in data:
    left_ids.update(set(episode['left_hand_touches']))
    right_ids.update(set(episode['right_hand_touches']))
    left_cumulative.append(len(left_ids))
    right_cumulative.append(len(right_ids))   
```

```python
plt.plot(left_cumulative, label='Left Hand')
plt.plot(right_cumulative, label='Right Hand')
plt.xlabel('Episode')
plt.ylabel('Unique Body Parts Touched')
plt.title('Cumulative Unique Body Parts Touched Over Episodes')
plt.legend()
plt.show()
```
