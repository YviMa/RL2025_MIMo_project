```python
import numpy as np
import matplotlib.pyplot as plt

tau_h = 0.4
tau_d = 1.5


x1 = np.linspace(1, 2, 500)
x2 = np.linspace(2, 5, 500)


y1 = np.exp(-(x1-1)/tau_h)
y2 = (1 - np.exp(-(x2-1.85)/tau_d))

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.ylabel('reward')
plt.xlabel('t')
#plt.xticks([])
#plt.yticks([])
```

```python
import numpy as np
import matplotlib.pyplot as plt

tau_h = 0.6
tau_d = 1.5

# x-Bereiche
x1 = np.linspace(1, 2, 500)

# y1 und Endwert
y1 = np.exp(-(x1 - 1) / tau_h)
y1_end = y1[-1]  # = exp(-1/tau_h)

# horizontale Verschiebung a so, dass y2(2+a) = y1_end
a = -tau_d * np.log(1 - y1_end)

# verschobene x-Achse fr y2
x2_shift = np.linspace(2 + a, 5 + a, 500)

# verschobene y2
y2_shift = 1 - np.exp(-(x2_shift - (2 + a)) / tau_d)

# Plot
plt.plot(x1, y1, label="y1")
plt.plot(x2_shift, y2_shift, label="y2 (rechts verschoben)")
plt.axvline(2 + a, linestyle=":", label=f"bergang x = {2 + a:.3f}")
plt.ylabel('reward')
plt.xlabel('t')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Zeitachse
t = np.linspace(0, 10, 500)

# Parameter
tau_h = 0.5   # schnelle Abnahme
tau_d = 3.0   # langsamere Erholung

# Modell: nach dem Touch fllt Signal schnell ab und erholt sich langsam
signal = np.exp(-t/tau_h) + (1 - np.exp(-t/tau_d))

plt.plot(t, signal, label="schematische Kurve")
plt.axvline(0, color="gray", linestyle="--", label="touch")
plt.xlabel("Zeit")
plt.ylabel("Level")
plt.title("Schematische Skizze (h Abfall, d Erholung)")
plt.legend()
plt.show()
```
